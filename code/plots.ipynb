```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
sims = ['TNG300-1', 'TNG300-2', 'TNG300-3','TNG100-1', 'TNG100-2', 'TNG100-3', 'TNG50-1', 'TNG50-2', 'TNG50-3']
```

```python
data = {}
for sim in sims:
    data[sim] = np.loadtxt('../data/summary_vcm_{}.dat'.format(sim))
    
abacus = {}
for i in range(20):
    abacus[i] = np.loadtxt('../../PeculiarMotion/data/summary/cmspeed_abacus_planck_00_box_{:02d}.dat'.format(i))
```

```python
for sim in sims:
    n = len(data[sim])
    plt.plot(np.sort(data[sim]), np.linspace(1/n, 1.0, n))

for i in range(20):
    n = len(abacus[i])
    plt.plot(np.sort(abacus[i]), np.linspace(1/n, 1.0, n))
```
